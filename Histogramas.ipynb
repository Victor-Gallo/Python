## Histogramas a partir de un Timeseries
Primero se debe importar las librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Luego, cargar el archivo que contiene la información.
El argumento "Index_col" se utiliza para indicar que columna será el indice, en este caso en la columna de posición cero (0)

```python
pm = pd.read_csv('D:/VICTOR/Manejo de Cuencas/Python_basico/pm_2.csv', index_col = 0, parse_dates=True)
pm
```

```python
Chalilla = pm['Chalilla'].plot(kind='line', figsize = (12,5))
plt.title('Chalilla')
plt.xlabel('Meses')
plt.ylabel('mm/mes')
```

Histogramas

Histogramas con subplots

```python
fig = plt.figure()
ax0 = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(2,2,4)
plt.subplots_adjust(wspace=0.3, hspace=0.5)


pm['CHALILLA'].plot(kind='line', figsize = (15,6), ax = ax0).set_title('Chalilla')
ax0.set_ylabel('mm/mes')
pm['SAN DAMIAN'].plot(kind='line', figsize = (15,6), ax = ax1).set_title('San Damian')
ax1.set_ylabel('mm/mes')
pm['LANGA'].plot(kind='line', figsize = (15,6), ax = ax2).set_title('Langa')
ax2.set_ylabel('mm/mes')
pm['ANTIOQUIA'].plot(kind='line', figsize = (15,6), ax = ax3).set_title('Antioquía')
ax3.set_ylabel('mm/mes')
```

```python
plt.style.use('seaborn')

fig = plt.figure()
ax0 = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(2,2,4)
plt.subplots_adjust(wspace=0.3, hspace=0.5)


pm['CHALILLA'].plot(kind='line', figsize = (15,6), ax = ax0).set_title('Chalilla')
ax0.set_ylabel('mm/mes')
pm['SAN DAMIAN'].plot(kind='line', figsize = (15,6), ax = ax1).set_title('San Damian')
ax1.set_ylabel('mm/mes')
pm['LANGA'].plot(kind='line', figsize = (15,6), ax = ax2).set_title('Langa')
ax2.set_ylabel('mm/mes')
pm['ANTIOQUIA'].plot(kind='line', figsize = (15,6), ax = ax3).set_title('Antioquía')
ax3.set_ylabel('mm/mes')
```

Crearemos una columna que tenga datos categoricos que se llame mes, en el dataframe pm que habíamos creado

```python
pm['mes'] = pm.index.month
```

```python
pm
```

```python
pm.boxplot(column = 'Chalilla', by='mes')
```

```python
plt.style.use('seaborn')

fig = plt.figure()
ax0 = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(2,2,4)
plt.subplots_adjust(wspace=0.3, hspace=0.5)

pm.boxplot(column = 'Chalilla', by='mes', figsize = (18,7), ax = ax0).set_title('Chalilla')
ax0.set_ylabel('mm/mes')
pm.boxplot(column = 'San Damian', by='mes', figsize = (18,7), ax = ax1).set_title('San Damian')
ax0.set_ylabel('mm/mes')
pm.boxplot(column = 'Langa', by='mes', figsize = (18,7), ax = ax2).set_title('Langa')
ax0.set_ylabel('mm/mes')
pm.boxplot(column = 'Antioquia', by='mes', figsize = (18,7), ax = ax3).set_title('Antioquia')
ax0.set_ylabel('mm/mes')
```

```python
sns.set(rc={"figure.figsize":(12,7)})
sns.boxplot( x=pm['mes'], y=pm['Chalilla']);
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(2,2,4)
plt.subplots_adjust(wspace=0.3, hspace=0.5)

sns.boxplot( x=pm['mes'], y=pm['Chalilla'], ax = ax0).set_title('Chalilla')
ax0.set_ylabel('mm/mes');
sns.boxplot( x=pm['mes'], y=pm['San Damian'], ax = ax1).set_title('San Damian')
ax0.set_ylabel('mm/mes');
sns.boxplot( x=pm['mes'], y=pm['Langa'], ax = ax2).set_title('Langa')
ax0.set_ylabel('mm/mes');
sns.boxplot( x=pm['mes'], y=pm['Antioquia'], ax = ax3).set_title('Antioquia')
ax0.set_ylabel('mm/mes');
```
